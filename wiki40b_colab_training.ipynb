# GRIM NLP Model - Google Colab Training

This notebook allows you to train the GRIM NLP model on Google Colab using the Hugging Face `wiki40b_ja` dataset.

### Steps:
1. Clone the GRIM repository from GitHub.
2. Install required dependencies.
3. Start training using the `scripts/train.py` script (leveraging GPU and mixed precision).
4. Run inference / text generation using the trained checkpoint.

```python
# 1. Clone the repository and change directory
!git clone https://github.com/YJSNPI114514e/Model.git
%cd Model
```

```python
# 2. Install dependencies (standard Google Colab environment has PyTorch pre-installed)
!pip install datasets torchdiffeq tqdm numpy gradio
```

```python
# 3. Start GRIM Language Model training
# We use streaming mode to load the giant wiki40b_ja dataset efficiently.
# Feel free to adjust --max-samples, --epochs, and --D (hidden dimension) as needed.
!python scripts/train.py \
    --dataset fn-aka-mur/wiki40b_ja \
    --max-samples 50000 \
    --streaming \
    --epochs 10 \
    --seq-len 64 \
    --D 128 \
    --device cuda \
    --amp
```

```python
# 4. Generate Text using the trained checkpoint
# You can test the generation quality using the best checkpoint.
!python scripts/generate.py \
    --checkpoint checkpoints/best.pt \
    --prompt "意識は場である。" \
    --max-new-tokens 50 \
    --temperature 0.8
```
